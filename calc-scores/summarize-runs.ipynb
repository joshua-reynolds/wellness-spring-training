```python
import os
import pandas as pd

# show all columns
pd.options.display.max_columns = None
```

```python
if not os.path.exists('Outputs'):
    os.makedirs('Outputs')
  
outputs = '.\\Outputs'
```

```python
week1 = pd.read_csv(r".\Inputs\teams1_2024.csv")[['team_id','name', 'runs']]
week2 = pd.read_csv(r".\Inputs\teams2_2024.csv")[['team_id', 'name', 'runs']]
week3 = pd.read_csv(r".\Inputs\teams3_2024.csv")[['team_id', 'name', 'runs']]
week4 = pd.read_csv(r".\Inputs\teams4_2024.csv")[['team_id', 'name', 'runs']]

bats1 = pd.read_csv(r".\Inputs\bats1_2024.csv").merge(week1, on='team_id', how='left')
bats2 = pd.read_csv(r".\Inputs\bats2_2024.csv").merge(week2, on='team_id', how='left')
bats3 = pd.read_csv(r".\Inputs\bats3_2024.csv").merge(week3, on='team_id', how='left')
bats4 = pd.read_csv(r".\Inputs\bats4_2024.csv").merge(week4, on='team_id', how='left')
```

```python
# need to join to respective week's teams file for names first, since ids vary
week1 = week1[['name', 'runs']].copy()
week2 = week2[['name', 'runs']].copy()
week3 = week3[['name', 'runs']].copy()
week4 = week4[['name', 'runs']].copy()

week1.columns = ['name', 'runs1']
week2.columns = ['name', 'runs2']
week3.columns = ['name', 'runs3']
week4.columns = ['name', 'runs4']
```

```python
# wpt = week1.merge(week2, on='name', how='left').merge(week3, on='name', how='left')
wpt = week1.merge(week2, on='name', how='left').merge(week3, on='name', how='left').merge(week4, on='name', how='left')
```

```python
# wpt['total_runs'] = wpt['runs1'] + wpt['runs2'] + wpt['runs3']
wpt['total_runs'] = wpt['runs1'] + wpt['runs2'] + wpt['runs3'] + wpt['runs4']
wpt
```

```python
bats1_summary = pd.pivot_table(bats1, values='bat_id', index=['name'], columns=['bat_result'], aggfunc="count").fillna(0).reset_index()
bats1_summary['total_at_bats'] = bats1_summary[0] +  bats1_summary[1] + bats1_summary[2] + bats1_summary[3] + bats1_summary[4]
bats1_summary.columns = ['name', 'outs_1', 'singles_1', 'doubles_1', 'triples_1', 'homeruns_1', 'total_at_bats1']

bats2_summary = pd.pivot_table(bats2, values='bat_id', index=['name'], columns=['bat_result'], aggfunc="count").fillna(0).reset_index()
bats2_summary['total_at_bats'] = bats2_summary[0] +  bats2_summary[1] + bats2_summary[2] + bats2_summary[3] + bats2_summary[4]
bats2_summary.columns = ['name', 'outs_2', 'singles_2', 'doubles_2', 'triples_2', 'homeruns_2', 'total_at_bats2']

bats3_summary = pd.pivot_table(bats3, values='bat_id', index=['name'], columns=['bat_result'], aggfunc="count").fillna(0).reset_index()
bats3_summary['total_at_bats'] = bats3_summary[0] +  bats3_summary[1] + bats3_summary[2] + bats3_summary[3] + bats3_summary[4]
bats3_summary.columns = ['name', 'outs_3', 'singles_3', 'doubles_3', 'triples_3', 'homeruns_3', 'total_at_bats3']

bats4_summary = pd.pivot_table(bats4, values='bat_id', index=['name'], columns=['bat_result'], aggfunc="count").fillna(0).reset_index()
bats4_summary['total_at_bats'] = bats4_summary[0] +  bats4_summary[1] + bats4_summary[2] + bats4_summary[3] + bats4_summary[4]
bats4_summary.columns = ['name', 'outs_4', 'singles_4', 'doubles_4', 'triples_4', 'homeruns_4', 'total_at_bats4']

wpt = (wpt.merge(bats1_summary, on='name', how='left')
           .merge(bats2_summary, on='name', how='left')
           .merge(bats3_summary, on='name', how='left')
           .merge(bats4_summary, on='name', how='left'))

wpt['total_at_bats'] = wpt['total_at_bats1'] + wpt['total_at_bats2'] + wpt['total_at_bats3'] + wpt['total_at_bats4']
wpt['total_homeruns'] = wpt['homeruns_1'] + wpt['homeruns_2'] + wpt['homeruns_3'] + wpt['homeruns_4']
wpt['total_outs'] = wpt['outs_1'] + wpt['outs_2'] + wpt['outs_3'] + wpt['outs_4']
wpt
```

```python
wpt.fillna(0).to_csv(os.path.join(outputs, 'summary_full.csv'), index=False)
```
